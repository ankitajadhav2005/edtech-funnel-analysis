```python
# count leads per cycle
cycle_counts = sales_manager_assigned_leads.groupby('cycle')['lead_id'].nunique().reset_index()


# rename for clarity
cycle_counts.columns = ['cycle', 'total_leads']

print(cycle_counts)
```

```python
plt.figure()
plt.bar(cycle_counts['cycle'], cycle_counts['total_leads'])
plt.xlabel("Cycle")
plt.ylabel("Total Leads")
plt.title("Total Leads Distributed Across Cycles")
plt.show()
```

```python
# cycle wise dates
dates = sales_manager_assigned_leads[['cycle', 'assigned_date']].drop_duplicates()
dates
```

```python
# days of difference between each cycle
dates['days_diff'] = dates['assigned_date'].diff().dt.days
dates['days_diff']
```

```python
plt.figure()
plt.plot(dates['cycle'], dates['days_diff'], marker='o')
plt.xlabel("Cycle")
plt.ylabel("Days Difference")
plt.title("Days Between Each Cycle")
plt.show()
```
